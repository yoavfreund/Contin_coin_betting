```python
%pylab inline
from scipy.special import gamma
import numpy as np
```

```python
gamma(4)
```

```python
def F(t,x):
    a=gamma((t+1.+x)/2)*gamma((t+1.-x)/2)
    b=(2**t)/(np.pi*gamma(t+1))
    return a*b
def beta(t,x):
    return (F(t,x+1)-F(t,x-1))/(F(t,x+1)+F(t,x-1))
```

```python
F(2,1),beta(2,1)
```

```python
X=np.arange(-3,3,0.01)
Y=np.zeros(len(X))
X.shape[0],Y.shape
```

```python
for t in range(3,5):
    for i in range(X.shape[0]):
        Y[i] = F(t,X[i])
    plot(X,Y,label=str(t))
legend()
```

```python
def upper(g,x,t):
    return (1+g*beta(t,x))*F(t-1,x)
```

```python
x=-1; t=100;
X=arange(-1,+1,0.01)
Fv=np.zeros(len(X))
Fv1=np.zeros(len(X))
U=np.zeros(len(X))
for i in range(X.shape[0]):
    Fv[i]=F(t,x+X[i])
    Fv1[i]=F(t-1,x+X[i])
    U[i]=upper(X[i],x,t)
    
```

```python
figure(figsize=[10,8])
plot(X,Fv,label='F(t)')
plot(X,Fv1,label='F(t-1)')
plot(X,U,label='upper bound')
#xlim([-0.3,0.3])
legend()
```
